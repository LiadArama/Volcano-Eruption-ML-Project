```python
import warnings
warnings.filterwarnings("ignore")
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline 
import pygal
from matplotlib import rcParams
```

```python
final_df = pd.read_csv('final-df.csv')
```

```python
final_df
```

```python
final_df.corr(method = 'spearman')#spearman did the best with corr.
```

```python
fig = plt.figure(figsize = (20,20))
ax = fig.gca()
final_df.hist(ax = ax,bins=50)
```

```python
rcParams['figure.figsize'] = 8,6
plt.scatter(final_df['Max. VEI'], final_df['Number Of Eruptions'],c=final_df['Max. VEI'],cmap = 'rainbow')
plt.title("Relation between Max VEI to Number Of Eruptions")
plt.show()
ax = sns.violinplot(x="Max. VEI", y="Number Of Eruptions", data=final_df)
sns.lineplot(x='Max. VEI', y='Number Of Eruptions', data=final_df, color="indigo")
```

```python
sns.lineplot(x='Year', y='Number Of Eruptions', data=final_df, color="indigo")
```
