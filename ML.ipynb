```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neighbors import NearestNeighbors
from sklearn.metrics import confusion_matrix
from sklearn.metrics import f1_score
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.neural_network import MLPClassifier
from sklearn.neural_network import MLPRegressor
```

```python
df = pd.read_csv("final-df.csv")
```

```python
df
```

```python
temp_cols=df.columns.tolist()#list for cols
index=df.columns.get_loc("Max. VEI")#target column switch places with 1st column
new_cols=temp_cols[index:index+1] + temp_cols[0:index] + temp_cols[index+1:]
df=df[new_cols]
df
```

```python
X = df.iloc[:, 1:10] #splitting the data to target column and the rest
y = df.iloc[:, 0]
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.2)#training 80% testing 20%
```

```python
def knn(X_train, y_train, X_test, y_test):#K nearest neighbors prediction model
    sc_X = StandardScaler()
    X_train = sc_X.fit_transform(X_train)
    X_test = sc_X.transform(X_test)
    
    classifier = KNeighborsClassifier(n_neighbors=7, p=2, metric='euclidean', n_jobs=-1)
    classifier.fit(X_train, y_train)
    y_pred = classifier.predict(X_test)
    return y_pred

def mlp(X_train, y_train, X_test, y_test): #MLP - nueron network model - found on scikit (great website!)
    clf = MLPClassifier().fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    return y_pred

```

```python
def evaluate(X_train, y_train, X_test, y_test, classifier):#evaluating our score
    y_pred = classifier(X_train, y_train, X_test, y_test)
    print("%.6f" % precision_score(y_test, y_pred, average='micro'))
    print("%.6f" % recall_score(y_test, y_pred, average='micro'))
    print("%.6f" % f1_score(y_test, y_pred, average='micro'))
    print("%.6f" % accuracy_score(y_test, y_pred))
    print()
    print(confusion_matrix(y_test, y_pred))
    print('\n============================================')

evaluate(X_train, y_train, X_test, y_test, knn)
evaluate(X_train, y_train, X_test, y_test, mlp)
```
